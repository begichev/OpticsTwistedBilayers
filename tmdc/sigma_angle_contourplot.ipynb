```python
import numpy as np
import matplotlib.pyplot as plt
import os
from itertools import compress

### system settings
def mkdir_p(mypath):
    from errno import EEXIST
    try:
        os.makedirs(mypath)
    except OSError as exc:
        if exc.errno == EEXIST and os.path.isdir(mypath):
            pass
        else: raise
###

path = '/Users/daria/mdirac_njit/out/kq_tWSe2_eg170/resigma36_cool'
outpath = path+'/resigma_contourplot'
mkdir_p(outpath)

sigm_dat = np.loadtxt('out/kq_tWSe2_eg170/resigma36_cool/sigma_lcp_room_twist_full0.5.txt')

folders = [f for f in os.listdir(path) if not (f.startswith('.') or f.startswith('r'))]
#                                              or f.endswith('4corr'))]
folders.sort()

angle_shape = len(folders)
sigma2d = np.zeros((angle_shape,1000))
angles = np.zeros((angle_shape))
energies = sigm_dat[:,0]

sigmas = os.listdir(path)

angle_shape = len(folders)
sigma2d = np.zeros((angle_shape,1000))
angles = np.zeros((angle_shape))
energies = sigm_dat[:,0]
iv,jv = np.meshgrid(angles,energies,indexing='ij')

count=0
for s in sorted(os.listdir(path))[1:]:
    sigma2d[count,:] = np.loadtxt(path+'/'+s)[:,1]
    angles[count] = float(s[25:28])
    count = count+1

left_cut = 300
right_cut = 480
iv,jv = np.meshgrid(angles,energies[left_cut:right_cut],indexing='ij')
fig, ax = plt.subplots(figsize=(9,9))
plt.title('ReSigma, WSe2 indirect KQ, V = 36 meV, cryo T', fontsize=15)
ax.contourf(iv,jv,sigma2d[:,left_cut:right_cut],cmap='RdGy',levels=20)
ax.set_xlabel('angle, degrees', fontsize=20)
ax.set_ylabel('Photon energy, eV', fontsize=20)
          
#ax.set_aspect('equal', adjustable='box')
#plt.colorbar()
import matplotlib
norm = matplotlib.colors.Normalize(vmin=0, vmax=np.max(sigma2d))
cbar = plt.colorbar(matplotlib.cm.ScalarMappable(norm=norm, cmap='RdGy'))
cbar.ax.set_facecolor('none')
save_to = outpath+'/ContourPlotSigma.png'
plt.savefig(outpath+'/resigma_wse2_36_cool.png',dpi=180)

plt.close()
```
